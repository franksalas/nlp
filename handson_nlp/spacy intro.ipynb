```python
# load the package


import spacy

nlp = spacy.load('en')
```

```python
# Reading the text / tokens

doc = nlp('The walls on the mall are totaly totally tall!')
```

```python
doc
```

# read a file

```python
my_file = open('data/grail.txt').read()
```

```python
doc_file =  nlp(my_file)
```

```python
doc_file
```

## Sentence tokens


```python
for sentence in doc_file.sents:
    print(sentence)
```

```python
for num, sentence in enumerate(doc_file.sents):
    print(f'{num}: {sentence}')
```

## Word Tokens
- splitting or segmenting the text into words
- `.text`

```python
doc
```

```python
# word tokens
for token in doc:
    print(token.text)
```

```python
# list of word tokens
[token.text for token in doc]
```

```python
# similary to splitting on spaces

doc.text.split(' ')
```

## more about words
